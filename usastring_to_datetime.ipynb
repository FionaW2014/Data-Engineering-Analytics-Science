```python
import pandas as pd
```

```python
def own_str_to_date(datestring):
    monthdict = {'January': 1,
                'February': 2,
                'March': 3,
                'April': 4,
                'May': 5,
                'June': 6,
                'July': 7,
                'August': 8,
                'September': 9,
                'October': 10,
                'November': 11,
                'December': 12}
    
    #delete the comma in the datestring
    clear_datestring = datestring.replace(",", "")
    
    #parse datestring by space into list
    datelist = clear_datestring.split(" ")
    
    #use the month string as a key to seach in monthdict dictionary to return the equivalent numeric month
    #reassign datelist indexed 0 to the equivalent numeric value
    new_datelist = [0,0,0]
    new_datelist[1]  = str(monthdict.get(datelist[0])) #month
    
    #convert the likely strings of day and year to integer
    new_datelist[0]  = str(datelist[2]) #year
    new_datelist[2] = str(datelist[1]) #day
    new_datestr = ''.join(new_datelist)
    
    #regroup the new_datestring in form of 01/01/2000, or MM/DD/YYYY
    #more info at: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html
    new_date_form = pd.to_datetime(new_datestr, format='%Y%m%d', errors='coerce').date()
    
    return new_date_form

```
